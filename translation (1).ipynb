```
tokenizer_obj = Tokenizer()
```

```
tokenizer_german = Tokenizer( filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n', lower=True, split=' ', char_level=False)
```

```
tokenizer_english = Tokenizer( filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n', lower=True, split=' ', char_level=False)
```

```
tokenizer_german.fit_on_texts(german)
```

```
tokenizer_english.fit_on_texts(english)
```

```
tokenizer_german.word_index
```

```
tokenizer_english.word_index
```

```
for word in tokenizer_english.word_index :
  if word in english_stop_words :
    tokenizer_english.word_index.pop(word)
```

```
german_index_seq = tokenizer_german.texts_to_sequences(german)
```

```
english_index_seq = tokenizer_english.texts_to_sequences(english)
```

```
german_vocab_size = len(tokenizer_german.word_index) + 1
```

```
english_vocab_size = len(tokenizer_english.word_index) + 1
```

```
german_vocab_size
```

```
english_vocab_size
```

```
padded_german = pad_sequences(german_index_seq, maxlen=250, padding='post', truncating='post')
padded_german
```

```
padded_english = pad_sequences(english_index_seq, maxlen=200, padding='post', truncating='post')
padded_english
```

```
padded_english = padded_english.reshape(*padded_english.shape, 1)
```

```
model = Sequential()
model.add(Embedding(german_vocab_size, 1024, input_length = padded_german.shape[1], mask_zero=True))
model.add(LSTM(1024))
model.add(RepeatVector(padded_english.shape[1]))
model.add(LSTM(1024, return_sequences=True))
model.add(TimeDistributed(Dense(english_vocab_size, activation='softmax')))
model.compile(optimizer='Adam', loss='sparse_categorical_crossentropy', metrics=['acc'])
model.summary()
```

```
model.fit(padded_german[:500], padded_english[:500], batch_size=2, validation_split=0.05)
```

```
index_to_words = {id: word for word, id in tokenizer_german.word_index.items()}
index_to_words[0] = " "
```
